#C_CONTIGUOUS means row major indexing of array.
#F_CONTIGUOUS means column major indexing of array.
#Let's take a example on both types of indexing.

```python
c_array=np.random.rand(10, 10)
f_array=np.asfortranarray(c_array)
#return the sum of zeroth row
def sum_row(x):
    return np.sum(x[0, :])
#return the sum of zeroth column
def sum_col(x):
    return np.sum(x[:, 0])

#the summing up the row of a C array is much faster than summing up its column because in C array elements in a row are laid in successive locations.
#the opposite is true for Fortran array where the elements of a column are laid in a consecutive memory locations.
%timeit sum_row(c_array)
%timeit sum_row(f_array)
%timeit sum_col(c_array)
%timeit sum_col(f_array)
```

# Views

```python
x=np.random.rand(100,10)
#here y is the view of x
y=x[:5, :]
#may_share_function determine whether two arrays are view or copy of each other
print(np.may_share_memory(x,y))
y[:]=0
print(x[:5, :])
```

# Copies 

```python
x=np.random.rand(100,10)
y=np.empty([5,10])
#here y is copy of x
y[:]=x[:5, :]
print(np.may_share_memory(x,y))
y[:]=0
print(x[:5, :])
```

# Vectorized Operations

```python
#all the elements of array added by 1 simutaneously
x=np.array([1,2,3,4])
print(x+1)

#two numpy arrays are mutipied element by element
y=np.array([-1,2,3,0])
print(x*y)

#matrix multiplication using dot func
print(np.dot(x, y))

#comparison between two arrays
print(x==y)
```

# Universal Function

Square Function

```python
x=np.arange(5,10)
print(np.square(x))
```

mod func

```python
y=np.ones(5)*10
print(np.mod(y,x))
```

minimum func - it compares two arrays and returns the minimum value for both

```python
print(np.minimum(x,7))
```

min func - it returns the smallest element in the array

```python
print(np.min(x))
```

repeat func - it repeats the given array n number of times and assign it to a new variable

```python
z=np.repeat(x, 3).reshape(5,3)
print(z)
```

median func

```python
print(np.median(z))
#axis=0 i.e. the operation is based on column
print(np.median(z, axis=0))
#axis=1 i.e. the operation is based on rows
print(np.median(z, axis=1))
```

accumulate func - to accumulate the result of applying add of all elements

```python
print(np.add.accumulate(x))
```

multiply outer func - it multiplies the all pairs of elements from two input arrays

```python
print(np.multiply.outer(x, x))
```

# Broadcasting Rules

```python
x=np.array([[0,0,0], [10,10,10], [20,20,20]])
y=np.array([1,2,3])
#the x variable has of (3,3), while y has a shape of 3. But in numpy broadcasting the shape of y is translated to 3 by 1
print(x+y)
```

# Reshaping

```python
#In this example we have an array of shape (24,1) and after altering the shape attribute the shape has been changed to 2 by 3 by 4
x=np.arange(24)
x.shape=2, 3, -1
print(x)
```

# Vector Stacking

It construct a two or multidimensional array by equally sized row vectors
#It has three types

```python
x=np.arange(0, 10, 2)
y=np.arange(0, -5, -1)
#vstack - It construct new array by vertical stacking two arrays
print(np.vstack([x, y]))
#hstack - It combines two arrays horizontally
print(np.hstack([x, y]))
#dstack - It stacks the arrays in sequence depth-wise along the third dimension
print(np.dstack([x, y]))
```

resize func

```python
x=np.arange(3)
print(np.resize(x, (8,)))
```
